Dropout Regularization : Mostly used in CV problems with complex Deep Neural Network with too many layers and neurons

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sonar_dataset.csv',header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
x = df.drop(60,axis=1)
y = df[60]
y.sample(3)
```

```python
x.sample(3)
```

```python
y = pd.get_dummies(y,drop_first=True)
y.sample(3)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape , x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'), # or use input_dim=60 for only 1 dimension
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,y_pred))
```

```python
# introducing dropout layer in the new model

model_2 = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'), # or use input_dim=60 for only 1 dimension
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

model_2.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy']
)

model_2.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model_2.evaluate(x_test,y_test)
```

```python
# little imporvement in accuracy has been observed 
```

```python
y_pred = model_2.predict(x_test).reshape(-1)
print(y_pred[:10])
y_pred = np.round(y_pred)
print(y_pred[:10])
print(classification_report(y_test,y_pred))
```
